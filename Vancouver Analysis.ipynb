````python
import re

# Step 1: 删除 41 条 'New' (如果已经删过这步会 no-op)
before = len(df)
df = df[df['Name'] != 'New'].reset_index(drop=True)
print(f"删除 {before - len(df)} 条 'New'，剩余 {len(df)} 行")


def categorize(name):
    """
    Hybrid 分类逻辑：
    - 优先级 1 (最高): 基于 Chrome 手工核查得到的关键词 override
    - 优先级 2: 基于 keyword regex 的自动规则
    - 优先级 3 (兜底): General Day Camp / Other/Special
    """
    if not isinstance(name, str) or name.strip() == '':
        return 'Other/Special'
    n = name.lower()

    # ============================================================
    # 手工 override —— 基于实际查看 camp 描述得到的规则
    # 这些 camp 名字中没有明显类别关键词，但经核查后有专项归属
    # ============================================================

    # Sports override（飞盘/匹克球/体操/武术/篮球专项/体能/瑜伽等）
    if any(kw in n for kw in [
        'ultimate frisbee', 'pickleball', 'game ready crazy sports',
        'high 5 sports', 'rhythmic gymnastics', 'recreational gymnastics',
        'self defence', 'future bounce', 'shooters club',
        'kinetic kids', 'mindfulness & movement'
    ]):
        return 'Sports'

    # STEM override（各类机器人品牌/太空科学/工程设计）
    if any(kw in n for kw in [
        'mindstorms', 'wedo', 'spike robotics', 'robotics 1',
        "tomorrow's playground", 'steam 4 kids', 'space blast off',
        'space explorer', 'space explorers', 'inventor',
        'design and build', 'steamers'
    ]):
        return 'STEM'

    # Arts — Creative override（卡通/角色设计/电影制作/艺术/烹饪/创意设计）
    if any(kw in n for kw in [
        'cartooning', 'character design', 'young moviemakers',
        'little artist', 'famous artists remix', 'creative remix',
        'creative cooks', 'little cooks', 'fabric camp',
        'stories in the studio', 'amusement parks camp',
        'world cup cities', 'escape artist'
    ]):
        return 'Arts — Creative'

    # Arts — Performing/Dance override（主题舞蹈/音乐律动）
    if any(kw in n for kw in [
        'bluey', 'peppa', 'wickedly good time',
        'mini performance', 'zumbini'
    ]):
        return 'Arts — Performing/Dance'

    # Other/Special override（学术/识字/leadership/public speaking/empowerment）
    if any(kw in n for kw in [
        'math & literacy', 'math and literacy', 'ready to write',
        'ready for kindergarten', 'pre-kindergarten',
        'ready for the bell', 'empowerment camp',
        'hummingbird', 'kingfisher',
        'public speaking', 'entrepreneur mindset',
        'mindset masters', 'money masters', 'leadership'
    ]):
        return 'Other/Special'

    # ============================================================
    # 自动规则
    # ============================================================

    if re.search(r'\bsupported\b|1:1|\bpro-?d\b', n):
        return 'Other/Special'

    if (re.search(r'\b(robot|coding|code|programming|scratch|minecraft|roblox|'
                  r'python|ai\b|stem|engineer|engineering|computer|tech|'
                  r'game\s?design|3d|inventor|invention|science|lab|chemistry|'
                  r'biology|physics|forensic|chess|efk|architect|architecture)\b', n)
        or 'lego' in n or 'brick' in n or 'construction' in n or 'byte camp' in n):
        return 'STEM'

    if re.search(r'\b(dance|ballet|hip[\s-]?hop|jazz|kpop|k-pop|zumb|tumble|'
                 r'acrobat|choreograph|popping|locking|breaking|theatre|'
                 r'theater|drama|musical|act(?:ing)?|sing|singer|improv|'
                 r'broadway|play\s?making|music|guitar|piano|drum|band|'
                 r'ukulele|keyboard|dj\b|song)\b', n):
        return 'Arts — Performing/Dance'

    if re.search(r'\b(art\b|arts\b|craft|draw|drawing|paint|pottery|clay|'
                 r'sculpt|ceramic|collage|photograph|animation|anime|manga|'
                 r'film|filmmaking|media|fashion|sewing|cook|cooking|chef|'
                 r'baking|bake|culinary|kulinary|food|cartoon|cartoons|writing|'
                 r'storytelling|author|story|fairy|fairytale|superhero|video)\b', n):
        return 'Arts — Creative'

    if re.search(r'\b(sport|basketball|soccer|baseball|football|hockey|tennis|'
                 r'volleyball|badminton|golf|cricket|rugby|lacrosse|athletic|'
                 r'skateboard|cycling|bmx|track|karate|judo|taekwondo|martial|'
                 r'boxing|fencing|climbing|parkour|ninja|sportball|swim|aqua|'
                 r'pool|diving|yoga|fitness|capoeira)\b', n):
        return 'Sports'

    if re.search(r'\b(outdoor|adventure|hike|hiking|kayak|canoe|paddle|sail|'
                 r'fishing|bushcraft|wilderness|nature|forest|trail|eco\b|'
                 r'environmental|ranger|rangers)\b', n):
        return 'Outdoor & Adventure'

    if re.search(r'\b(day\s?camp|summer\s?camp|camp\b|funseekers|kids\s?camp|'
                 r'preschool|explorer|hangout|before\s?care|after\s?care|'
                 r'recreation|rec\b|daycamp|safari|smiles|daze|sunseekers|'
                 r'sunsplash|glow)\b', n):
        return 'General Day Camp'

    return 'Other/Special'


df['Category'] = df['Name'].apply(categorize)
print("\n=== 最终分类分布 ===")
print(df['Category'].value_counts())
print(f"\nTotal: {len(df)}")
````

````python
import re

# Step 1: 从 Age 字符串提取 min_age
def parse_min_age(age_str):
    if not isinstance(age_str, str):
        return None
    s = age_str.lower().strip()
    # "Under X yrs" → min = 0
    if s.startswith('under'):
        return 0
    # "X-Y yrs" 或 "X yrs" → 取前面的数字
    m = re.match(r'(\d+)', s)
    if m:
        return int(m.group(1))
    return None

df['Min Age'] = df['Age'].apply(parse_min_age)

# Step 2: 按 Min Age 分桶（3 档）
def assign_age_group(min_age):
    if pd.isna(min_age):
        return 'Unknown'
    if min_age <= 5:
        return 'Pre-K (0-5)'
    if min_age <= 9:
        return 'Child (6-9)'
    return 'Pre-teen/Teen (10+)'

df['Age Group'] = df['Min Age'].apply(assign_age_group)

# 查看结果
print("=== Age Group 分布（3 档，按 Min Age）===")
print(df['Age Group'].value_counts())
print(f"\nTotal: {len(df)}")
print(f"Min Age 缺失: {df['Min Age'].isna().sum()}")
````

## 👶 Age Group 分类说明（按 Min Age 分桶）

本项目采用 **Min Age（最低参加年龄）** 作为年龄分桶依据。
原因：家长选 camp 的核心决策是"我家孩子够不够格参加"——min age 是准入门槛，
直接驱动报名决策，比 max age 或年龄跨度更有业务信号。

**Pre-K (0-5)** — `min_age ≤ 5`  
学前儿童 camp。包括 0-4 岁的亲子音乐课（Music with Marnie, Zumbini）、
Pre-Kindergarten 学前衔接 camp、Preschool Summer Camp 等。
对应加拿大 Kindergarten 入学前阶段。

**Child (6-9)** — `6 ≤ min_age ≤ 9`  
小学低年级主力客群。对应加拿大 Elementary Junior（K-3）阶段。
这是暑期 camp 市场的最大客群，家长暑期托管需求最旺盛。

**Pre-teen/Teen (10+)** — `min_age ≥ 10`  
小学高年级 + 初中 camp。对应加拿大 Elementary Senior (4-7) 到
Secondary 早期阶段。这个群体通常更独立，camp 内容也更专项化
（高阶运动训练、编程、Leadership 等）。
