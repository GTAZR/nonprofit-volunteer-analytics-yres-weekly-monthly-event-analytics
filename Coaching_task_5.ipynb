```python
# 需要对2026年时间做转化！
df_2026['Start Date & Time'] = pd.to_datetime(df_2026['Start Date & Time'], errors = 'coerce')
df_2026['End Date & Time'] = pd.to_datetime(df_2026['End Date & Time'], errors = 'coerce')
```

```python
df_2025.columns = df_2025.columns.str.strip().str.capitalize()
df_2026.columns = df_2026.columns.str.strip().str.capitalize()
```

```python
df = pd.concat([df_2025, df_2026], ignore_index = True)
```

```python
# 检查重复项并删除！
print(df_2025.duplicated().sum())
print(df_2026.duplicated().sum())
print(df.duplicated().sum())
```

```python
df_cleaned = df.drop_duplicates().reset_index(drop=True)
```

```python
df_cleaned[['Province', 'Program name']] = df_cleaned['Event type name'].str.split(':', n = 1, expand = True)
```

```python
df_cleaned['Province'] = df_cleaned['Province'].str.strip()
df_cleaned['Program name'] = df_cleaned['Program name'].str.strip()
```

```python
# 一周代表从周日到周六
df_cleaned['Week Range'] = df_cleaned['Start date & time'].dt.to_period('W-SAT').apply(lambda r:f"{r.start_time.date()} to {r.end_time.date()}")
```
