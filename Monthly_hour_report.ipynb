### 目前最具有挑战性的是Date Joined VSP这一列的制作！
### 方法是：将 VSP 表拆成两个小表（分别对应两个 Email 列），分别与 final_report 匹配，最后再把结果合在一起

```python
# 1. 准备 VSP 的两个子表，只取关键列
vsp_subset1 = vsp[['Name', 'YRES Email', 'Creation log']].rename(columns={'Name': 'Full name', 'YRES Email': 'Email', 'Creation log': 'Date joined VSP'})
vsp_subset2 = vsp[['Name', 'Email Address', 'Creation log']].rename(columns={'Name': 'Full name', 'Email Address': 'Email', 'Creation log': 'Date joined VSP'})

# 2. 把这两个子表垂直拼在一起，形成一个“姓名-任意邮箱”的对照表，并去重
vsp_mapping = pd.concat([vsp_subset1, vsp_subset2]).dropna(subset=['Email']).drop_duplicates(subset=['Full name', 'Email'])

# 3. 将这个对照表合并到 final_report 中
final_report = pd.merge(final_report, vsp_mapping, on=['Full name', 'Email'], how='left')
```

```python
final_report
```

```python
final_report['Total Hours'] = final_report['Service'] + final_report['Training']
final_report
```

```python
# 1. 准备 VSP 的两个子表，只取关键列
vsp_subset3 = vsp[['Name', 'YRES Email', 'Youth Advisor 3 - Winter 2026']].rename(columns={'Name': 'Full name', 'YRES Email': 'Email', 'Youth Advisor 3 - Winter 2026': 'Youth Advisor 3 - Winter 2026'})
vsp_subset4 = vsp[['Name', 'Email Address', 'Youth Advisor 3 - Winter 2026']].rename(columns={'Name': 'Full name', 'Email Address': 'Email', 'Youth Advisor 3 - Winter 2026': 'Youth Advisor 3 - Winter 2026'})

# 2. 把这两个子表垂直拼在一起，形成一个“姓名-任意邮箱”的对照表，并去重
vsp_mapping_2 = pd.concat([vsp_subset3, vsp_subset4]).dropna(subset=['Email']).drop_duplicates(subset=['Full name', 'Email'])

# 3. 将这个对照表合并到 final_report 中
final_report = pd.merge(final_report, vsp_mapping_2, on=['Full name', 'Email'], how='left')
```

### 增加Sum of total hours by Feb 28以及Sum of total hours by Mar 31这两列！

```python
# cacculate the total hours by 2026.2.28
feb_cutoff = log[log['Volunteering Date'] <= '2026-02-28'].groupby(['Full name', 'Email'])['Total Volunteering Hours Completed Today'].sum().reset_index()
feb_cutoff.rename(columns={'Total Volunteering Hours Completed Today': 'Sum of Total hours by Feb 28, 2026'}, inplace = True)
```

```python
# 2. 计算截至 2026-03-31 的总时长
mar_cutoff = log[log['Volunteering Date'] <= '2026-03-31'].groupby(['Full name', 'Email'])['Total Volunteering Hours Completed Today'].sum().reset_index()
mar_cutoff.rename(columns={'Total Volunteering Hours Completed Today': 'Sum of Total hours by Mar 31, 2026'}, inplace=True)
```

```python
feb_cutoff
```

```python
mar_cutoff
```

```python
# 3. 合并到 final_report
final_report = pd.merge(final_report, feb_cutoff, on=['Full name', 'Email'], how='left').fillna(0)
final_report = pd.merge(final_report, mar_cutoff, on=['Full name', 'Email'], how='left').fillna(0)
```

```python
final_report
```

```python
import numpy as np

# 定义判断条件
conditions = [(final_report['Sum of Total hours by Mar 31, 2026'] >= 40) & (final_report['Sum of Total hours by Feb 28, 2026'] < 40),
             (final_report['Sum of Total hours by Mar 31, 2026'] >= 80) & (final_report['Sum of Total hours by Feb 28, 2026'] < 80),
              (final_report['Sum of Total hours by Mar 31, 2026'] >= 120) & (final_report['Sum of Total hours by Feb 28, 2026'] < 120)
]

choices = ['Reached 40h', 'Reached 80h','Reached 120h']

final_report['Milestone in March 2026'] = np.select(conditions, choices, default='')
```
